```python
import numpy as np
import matplotlib.pyplot as plt
import pylab
%matplotlib inline
import numpy.random as rd # importing random number package
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D # importing 3d plot package
```

```python
#CUBE UNIVERSE UNIFORM DISTRIBUTION, DISCARDING TO FORM A SPHERE
# making random numbers with python:

rd.seed(5) # setting a seed - yours can be a different number

r1 = rd.uniform(-1, 1, 3000) #randomly generate 3000 uniformly distributed values between -1 and 1
#print(r1)
r2 = rd.uniform(-1, 1, 3000)
#print(r2)
r3 = rd.uniform(-1, 1, 3000)
#print(r3)
#create a list for radius of a sphere
r=np.sqrt(((r1**2)+(r2**2)+(r3**2)))

#Generates a uniformly distributed cube from r1 r2 r3,


#boolean test, create a true-false list corresponding to r values that fufile <=1, 
rtest=r<=1
#create array of acceptable r by applying boolean array and choose first 1000 elements
acceptr=(r[rtest][0:1000])
print(len(acceptr))
print ("\n values of acceptable radius: \n")
print(acceptr)
#print (r[rtest])
#now multiply our corresponding 0,1 or boolean list to extract the corresponding coords
#as boolean same size with same elements in r1,2,3 list correspond to same element in r and boolean
r1new=(r1[rtest][0:1000])
r2new=(r2[rtest][0:1000])
r3new=(r3[rtest][0:1000])
rnew=np.sqrt(((r1new**2)+(r2new**2)+(r3new**2)))
print( "%d data points within the spherical universe" % len(rnew)) # to ensure we have 1000 points to plot 
```

```python
# how to make a 3d plot when you have 3 arrays: 

fig = plt.figure()
fig = plt.figure(figsize=(8,8))
ax = fig.add_subplot(111, projection='3d', elev=30., azim=480.) # elev and azim can be changed for different view
ax.scatter(r1new, r2new, r3new, s=10,  c='b',alpha=0.4, depthshade=True)
ax.set_aspect('equal') #set axis equal
#title and axis labels
plt.title ('Uniform Spherical Universe of Unit radius with 1000 Stars')
plt.xlabel('X position (r1)')
plt.ylabel('Y position (r2)')
ax.set_zlabel('Z position (r3)')


```

```python
gaus=r1+r2+r3
plt.hist(gaus, bins=np.arange(min(gaus), max(gaus) + 0.05, 0.1))
plt.title ('Gausian / Irwin-Hall Diustribution of summation of spatial values\n (Central Limit Theorem)')
plt.xlabel('Radial length (r)')
plt.ylabel('Count / Frequency')
print('as the sample size tends to infinity, the profile \nmatches a Gausian Distribution')
```

```python
#z=rd.power(2, size=1000)
#L=10/z
#print(L)
L = np.ones(1000, dtype=float)*10 #creates an array of value 10 of length 1000 
b=L/(acceptr**2)
#print(b)
#print(len(b))
test2 = b >=100 #boolean again
bnew=(b[test2])
rnew=(acceptr[test2])

plt.figure()
plt.title (' Uniform Spherical Universe of Unit radius with 1000 points \n')
plt.xlabel('X position (r1)',alpha=0.3)
plt.ylabel('Y position (r2)')
plt.plot(acceptr,b,'ro')
plt.plot(rnew,bnew,'bo')
plt.show()
```
