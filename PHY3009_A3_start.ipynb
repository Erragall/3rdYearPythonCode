```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import numpy.random as rd # importing random number package
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D # importing 3d plot package
```

```python
# making random numbers with python:

rd.seed(10000) # setting a seed - yours can be a different number

r1 = rd.uniform(-10, 10, 20)
print(r1)
r2 = rd.normal(10, 2, size=20)
print(r2)
r3 = rd.power(5, size=20)
print(r3)
```

```python
# how to make a 3d plot when you have 3 arrays: 

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d', elev=20., azim=30.) # elev and azim can be changed for different view
ax.scatter(r1, r2, r3, s=4, c=None, depthshade=True)
```
